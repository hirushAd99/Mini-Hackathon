```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define rainfall bins
bins = [0, 25, 60, float('inf')]  # Adjust the bin edges
labels = ['0-25', '25-50', '50+']  # Adjust the labels

# Create a new column 'rainfall_range' with the corresponding labels
dfOI['number_of_skus_sold_range'] = pd.cut(dfOI['number_of_skus_sold'], bins=bins, labels=labels, right=False)

# Plot a bar chart to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(dfOI['number_of_skus_sold_range'], dfOI['region'], color='green', alpha=0.5)
plt.xlabel('number_of_skus_sold_range')
plt.ylabel('region')
plt.title('number_of_skus_sold_range vs. region')
plt.grid(True)
plt.show()
```
